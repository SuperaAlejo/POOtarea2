```python
from enum import Enum

class TipoPlaneta(Enum):
    TERRESTRE = 1
    GASEOSO = 2
    ENANO = 3

class Planeta:
    def __init__(self, nombre=None, cantidad_satelites=0, masa_kg=0.0,volumen_km3=0.0, diametro_km=0, distancia_al_sol_millones_km=0, tipo=TipoPlaneta.TERRESTRE, observable_simple_vista=False):
        self.nombre = nombre
        self.cantidad_satelites = cantidad_satelites
        self.masa_kg = masa_kg
        self.volumen_km3 = volumen_km3
        self.diametro_km = diametro_km
        self.distancia_al_sol_millones_km = distancia_al_sol_millones_km
        self.tipo = tipo
        self.observable_simple_vista = observable_simple_vista

    def imprimir(self):
        print(f"Nombre: {self.nombre}")
        print(f"Cantidad de satélites: {self.cantidad_satelites}")
        print(f"Masa (kg): {self.masa_kg}")
        print(f"Volumen (km³): {self.volumen_km3}")
        print(f"Diámetro (km): {self.diametro_km}")
        print(f"Distancia al Sol (millones de km): {self.distancia_al_sol_millones_km}")
        print(f"Tipo: {self.tipo.name}")
        print(f"Observable a simple vista: {'Sí' if self.observable_simple_vista else 'No'}")

    def densidad(self):
        if self.volumen_km3 == 0:
            return 0
        return self.masa_kg / self.volumen_km3
    
    def es_exterior(self):
        distancia_media_ua = (self.distancia_al_sol_millones_km * 1e6) / 149597870

        if distancia_media_ua > 3.4:
            return True
        else:
            return False
        

if __name__ == "__main__":
    tierra = Planeta(
        nombre="Tierra",
        cantidad_satelites=1,
        masa_kg=5.972e24,
        volumen_km3=1.08321e12,
        diametro_km=12742,
        distancia_al_sol_millones_km=149,
        tipo=TipoPlaneta.TERRESTRE,
        observable_simple_vista=True
    )

    jupiter = Planeta(
        nombre="Júpiter",
        cantidad_satelites=95, 
        masa_kg=1.898e27,
        volumen_km3=1.43128e15,
        diametro_km=139820,
        distancia_al_sol_millones_km=778,
        tipo=TipoPlaneta.GASEOSO,
        observable_simple_vista=True
    )

    tierra.imprimir()
    print(f"Densidad de la Tierra: {tierra.densidad()} kg/km³")
    print(f"¿La Tierra es un planeta exterior? {'Sí' if tierra.es_exterior() else 'No'}")
    
    print()
    
    jupiter.imprimir()
    print(f"Densidad de Júpiter: {jupiter.densidad()} kg/km³")
    print(f"¿Júpiter es un planeta exterior? {'Sí' if jupiter.es_exterior() else 'No'}")
```

```python
from enum import Enum

class TipoCombustible(Enum):
    GASOLINA = 1
    BIOETANOL = 2
    DIESEL = 3
    BIODIESEL = 4
    GAS = 5

class TipoAutomovil(Enum):
    CIUDAD = 1
    SUBCOMPACTO = 2
    COMPACTO = 3
    FAMILIAR = 4
    EJECUTIVO = 5
    SUV = 6

class Color(Enum):
    BLANCO = 1
    NEGRO = 2
    ROJO = 3
    NARANJA = 4
    AMARILLO = 5
    VERDE = 6
    AZUL = 7
    VIOLETA = 8

class Automovil:
    def __init__(self, marca:str, modelo:str, motor:int, tipo_combustible:TipoCombustible, tipo_automovil:TipoAutomovil, 
                    cantidad_puertas:int, cantidad_asientos:int, velocidad_maxima:int, color:Color, velocidad_actual:int=0):
        self.marca = marca
        self.modelo = modelo
        self.motor = motor
        self.tipo_combustible = tipo_combustible
        self.tipo_automovil = tipo_automovil
        self.cantidad_puertas = cantidad_puertas
        self.cantidad_asientos = cantidad_asientos
        self.velocidad_maxima = velocidad_maxima
        self.color = color
        self.velocidad_actual = velocidad_actual

    def getMarca(self):
        return self.marca
    
    def getModelo(self):
        return self.modelo
    
    def getMotor(self):
        return self.motor
    
    def getTipoCombustible(self):
        return self.tipo_combustible
    
    def getTipoAutomovil(self):
        return self.tipo_automovil
    
    def getCantidadPuertas(self):
        return self.cantidad_puertas
    
    def getCantidadAsientos(self):
        return self.cantidad_asientos
    
    def getVelocidadMaxima(self):
        return self.velocidad_maxima
    
    def getColor(self):
        return self.color
    
    def getVelocidadActual(self):
        return self.velocidad_actual
    
    def setMarca(self, marca:str):
        self.marca = marca

    def setModelo(self, modelo:str):
        self.modelo = modelo

    def setMotor(self, motor:int):
        self.motor = motor

    def setTipoCombustible(self, tipo_combustible:TipoCombustible):
        self.tipo_combustible = tipo_combustible

    def setTipoAutomovil(self, tipo_automovil:TipoAutomovil):
        self.tipo_automovil = tipo_automovil

    def setCantidadPuertas(self, cantidad_puertas:int):
        self.cantidad_puertas = cantidad_puertas

    def setCantidadAsientos(self, cantidad_asientos:int):
        self.cantidad_asientos = cantidad_asientos

    def setVelocidadMaxima(self, velocidad_maxima:int):
        self.velocidad_maxima = velocidad_maxima

    def setColor(self, color:Color):
        self.color = color

    def setVelocidadActual(self, velocidad_actual:int):
        self.velocidad_actual = velocidad_actual

    def acelerar(self, incremento:int):
        if self.velocidad_actual + incremento <= self.velocidad_maxima:
            self.velocidad_actual += incremento
        else:
            print("No se puede acelerar más allá de la velocidad máxima.")

    def desacelerar(self, decremento:int):
        if self.velocidad_actual - decremento >= 0:
            self.velocidad_actual -= decremento
        else:
            print("No se puede desacelerar más allá de 0 km/h.")

    def frenar(self):
        self.velocidad_actual = 0

    def tiempoLlegada(self, distancia:int):
        if self.velocidad_actual == 0:
            print("El automóvil está detenido. No se puede calcular el tiempo de llegada.")
            return None
        else:
            tiempo = distancia / self.velocidad_actual
            return tiempo

    def imprimir(self):
        print(f"Marca: {self.marca}")
        print(f"Modelo: {self.modelo}")
        print(f"Motor: {self.motor} cc")
        print(f"Tipo de combustible: {self.tipo_combustible.name}")
        print(f"Tipo de automóvil: {self.tipo_automovil.name}")
        print(f"Cantidad de puertas: {self.cantidad_puertas}")
        print(f"Cantidad de asientos: {self.cantidad_asientos}")
        print(f"Velocidad máxima: {self.velocidad_maxima} km/h")
        print(f"Color: {self.color.name}")
        print(f"Velocidad actual: {self.velocidad_actual} km/h")
        print()


if __name__ == "__main__":
    auto1 = Automovil(
        marca="Toyota",
        modelo="Corolla",
        motor=1800,
        tipo_combustible=TipoCombustible.GASOLINA,
        tipo_automovil=TipoAutomovil.CIUDAD,
        cantidad_puertas=4,
        cantidad_asientos=5,
        velocidad_maxima=180,
        color=Color.BLANCO)
    
    auto1.imprimir()
    auto1.setVelocidadActual(100)

    print("La velocidad actual del automóvil es:", auto1.getVelocidadActual(), "km/h")
    auto1.acelerar(20)
    print("La velocidad actual del automóvil es:", auto1.getVelocidadActual(), "km/h")
    auto1.desacelerar(50)
    print("La velocidad actual del automóvil es:", auto1.getVelocidadActual(), "km/h")
    auto1.frenar()
    print("La velocidad actual del automóvil es:", auto1.getVelocidadActual(), "km/h")
```

```python
class Circulo:
    def __init__(self, radio:float):
        self.radio = radio

    def area(self) -> float:
        return 3.14159 * (self.radio ** 2)

    def perimetro(self) -> float:
        return 2 * 3.14159 * self.radio

class Rectangulo:
    def __init__(self, base:float, altura:float):
        self.base = base
        self.altura = altura

    def area(self) -> float:
        return self.base * self.altura

    def perimetro(self) -> float:
        return 2 * (self.base + self.altura)
    
class Cuadrado:
    def __init__(self, lado:float):
        self.lado = lado

    def area(self) -> float:
        return self.lado ** 2

    def perimetro(self) -> float:
        return 4 * self.lado
    
class TrianguloRectangulo: #Suponemos que la base y la altura corresponden a los catetos
    def __init__(self, base:float, altura:float):
        self.base = base
        self.altura = altura

    def area(self) -> float:
        return (self.base * self.altura) / 2
    
    def hipotenusa(self) -> float:
        return (self.base ** 2 + self.altura ** 2) ** 0.5
    
    def perimetro(self) -> float:
        return self.base + self.altura + self.hipotenusa()
    
    def tipo(self) -> str: #Es obvio que un triángulo rectángulo no puede ser equilátero. Debe ser isósceles o escaleno.
        if self.base == self.altura: #La hipotenusa siempre será mayor que los catetos, por lo que sólo hay que comparar los catetos
            return "Isósceles"
        else:
            return "Escaleno"   
        

if __name__ == "__main__":
    circulo = Circulo(2)
    rectangulo = Rectangulo(1,2)
    cuadrado = Cuadrado(3)
    triangulo = TrianguloRectangulo(3,5)

    print("Área del círculo:", circulo.area())
    print("Perímetro del círculo:", circulo.perimetro())
    print()
    print("Área del rectángulo:", rectangulo.area())
    print("Perímetro del rectángulo:", rectangulo.perimetro())
    print()
    print("Área del cuadrado:", cuadrado.area())
    print("Perímetro del cuadrado:", cuadrado.perimetro())
    print()
    print("Área del triángulo rectángulo:", triangulo.area())
    print("Perímetro del triángulo rectángulo:", triangulo.perimetro())
    print("Hipotenusa del triángulo rectángulo:", triangulo.hipotenusa())
    print("Tipo del triángulo rectángulo:", triangulo.tipo())


    
```

```python
from enum import Enum

class TipoCuenta(Enum):
    AHORRO = 1
    CORRIENTE = 2

class CuentaBancaria:
    def __init__(self, nombre_titular:str, apellido_titular:str, no_cuenta:str, tipo:TipoCuenta, saldo:float=0.0):
        self.nombre_titular = nombre_titular
        self.apellido_titular = apellido_titular
        self.no_cuenta = no_cuenta
        self.tipo = tipo
        self.saldo = saldo

    def consultar_saldo(self) -> None:
        print(f"El saldo actual es: {self.saldo}")
 
    def consignar(self, cantidad:float) -> bool:
        if cantidad > 0:
            self.saldo += cantidad
            print(f"Se han consignado {cantidad} a la cuenta.")
            print(f"El nuevo saldo es: {self.saldo}")
            print()
            return True
        else:
            print("La cantidad a consignar debe ser mayor que 0.")
            print()
            return False

    def retirar(self, cantidad:float) -> bool:
        if cantidad > 0 and cantidad <= self.saldo:
            self.saldo -= cantidad
            print(f"Se han retirado {cantidad} de la cuenta.")
            print(f"El nuevo saldo es: {self.saldo}")
            print()
            return True
        else:
            print("La cantidad a retirar debe ser mayor que 0 y menor o igual al saldo actual.")
            print()
            return False
    
    def imprimir(self) -> None:
        print(f"Nombre del titular: {self.nombre_titular} {self.apellido_titular}")
        print(f"Número de cuenta: {self.no_cuenta}")
        print(f"Tipo de cuenta: {self.tipo.name}")
        print(f"Saldo actual: {self.saldo}")
        print()

if __name__ == "__main__":
    cuenta1 = CuentaBancaria("Juan", "Pérez", "123456789", TipoCuenta.AHORRO)

    cuenta1.imprimir()
    cuenta1.consignar(200000)
    cuenta1.consignar(300000)
    cuenta1.retirar(400000)
```
